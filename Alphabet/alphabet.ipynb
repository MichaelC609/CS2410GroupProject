```python
import os
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from tensorflow.keras import backend as K
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import LabelBinarizer
```

```python
train = pd.read_csv('/content/sample_data/sign_mnist_train.csv')
test = pd.read_csv('/content/sample_data/sign_mnist_test.csv') #not using rn
```

```python
train.head()
```

```python
# Seperate labels and pixel values
labels = train['label'].values
pixel_data = train.drop(columns=['label']).values # Remove the label column and get the rest of the remaining data (the images) - Tony
```

```python
pixel_data_normalized = pixel_data / 255.0 # Normalize pixel values to the range [0, 1]
```

```python
images = pixel_data_normalized.reshape(pixel_data_normalized.shape[0], 28, 28, 1) # (num images, 28x28 pixels, color channel (1 for grayscale)
```

```python
label_binrizer = LabelBinarizer()
labels = label_binrizer.fit_transform(labels)
```

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=(3,3), activation = 'relu', input_shape=(28, 28 , 1) ))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Conv2D(64, kernel_size = (3, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Conv2D(64, kernel_size = (3, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Flatten())
model.add(Dense(128, activation = 'relu'))

model.add(Dense(24, activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])

print(model.summary())
```

```python
# train model
history = model.fit(images, labels, epochs = 10, batch_size = 64)
```

```python
test_data = pd.read_csv('/content/sample_data/sign_mnist_test.csv')

#separate labels and images
test_labels = test_data['label'].values
test_images = test_data.drop(columns=['label']).values

#Reshape images
test_images = test_images.reshape(-1, 28, 28, 1)

# Normalize pixel values
test_images = test_images / 255.0
```

```python
test_loss, test_accuracy = model.evaluate(test_images, test_labels, verbose = 2)
print(f"Test Loss: {test_loss}, Test Accuracy: {test_accuracy}")
```
